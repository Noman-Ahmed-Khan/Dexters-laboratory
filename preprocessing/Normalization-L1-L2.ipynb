### **L1 and L2 Normalization Explained**

Normalization is a process used in machine learning to adjust feature values in a dataset so that they are on the same scale. This ensures that no single feature dominates the model due to its magnitude. L1 and L2 normalization are two common types of normalization, and each follows a different mathematical approach.

---

### **Key Concepts in L1 and L2 Normalization**

1. **Purpose**:
   - To make the model training process stable and efficient.
   - To ensure all features contribute equally to the model's predictions.

2. **How It Works**:
   - **L1 Normalization**: Scales the feature values so that the **sum of their absolute values** equals 1.
     \[
     x_{\text{normalized}} = \frac{x_i}{\sum |x_i|}
     \]
   - **L2 Normalization**: Scales the feature values so that the **square root of the sum of their squared values** equals 1.
     \[
     x_{\text{normalized}} = \frac{x_i}{\sqrt{\sum x_i^2}}
     \]

---

### **When to Use L1 vs. L2 Normalization**
- **L1 Normalization**:
  - Useful when features are sparse (most values are zero).
  - Ensures that only a few features dominate the scale.
  
- **L2 Normalization**:
  - Commonly used for dense datasets (most values are nonzero).
  - Ensures all features are scaled proportionally to their magnitude.

---


---

### **Conclusion**
- L1 normalization focuses on scaling values such that their absolute sum equals 1.
- L2 normalization focuses on scaling values so their vector magnitude equals 1.
- Both methods help to scale features and ensure that no feature dominates due to its scale, improving model performance and convergence during training.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
df=pd.DataFrame({'col01':np.random.exponential(size=1000,scale=2),
                 'col02':np.random.exponential(size=1000,scale=2),
                 'col03':np.random.exponential(size=1000,scale=2),
                 'col04':np.random.exponential(size=1000,scale=2),
                 'col05':np.random.exponential(size=1000,scale=2)})

total_positions = df.size  
random_positions = np.random.choice(total_positions, size=500, replace=False)

for pos in random_positions:
    row, col = divmod(pos, df.shape[1])
    df.iat[row, col] = 0

sns.histplot(df,alpha=0.5,kde=True)
plt.show()
```

```python
from sklearn.preprocessing import Normalizer

normalizer=Normalizer(norm='l1')
transformed_L1=normalizer.fit_transform(df)
sns.histplot(transformed_L1.flatten(),kde=True)
plt.show()
print(np.sum(transformed_L1,axis=1))
```

```python
transformed_L2=normalizer.fit_transform(df)
sns.histplot(transformed_L2.flatten(),kde=True)
plt.show()
print(pow(np.sum(transformed_L2,axis=1),2))
```

Normalization (like L1, L2, or max normalization) may still be needed after transformations in certain scenarios. Here’s a comprehensive list of cases where such normalization is beneficial:

---

### **1. When Algorithms Require Normalized Data**
Certain machine learning algorithms are sensitive to the scale or magnitude of features, even after transformations. Normalization ensures consistent scaling:
- **Algorithms involving distances**:
  - k-Nearest Neighbors (k-NN)
  - Support Vector Machines (SVM) (especially with radial kernels)
  - Clustering algorithms like K-Means
  - Principal Component Analysis (PCA)
- **Regularization-based models**:
  - Lasso Regression (L1 regularization)
  - Ridge Regression (L2 regularization)
  - Elastic Net
  - Neural Networks (gradient descent converges more efficiently with normalized inputs)

---

### **2. When Features Have Been Transformed but Still Vary in Scale**
Even after transformations like quantile or log, features might retain different scales or ranges. Normalization ensures that:
- All features contribute equally to the model.
- No feature dominates due to its magnitude.

---

### **3. When Working with Sparse Data**
- Sparse datasets (e.g., text data in TF-IDF format) often benefit from **L1 normalization** to ensure that:
  - The resulting vectors have comparable magnitudes.
  - The sparsity of the data is preserved.

---

### **4. When Combining Features with Different Transformations**
If your data includes:
- Some features that were transformed (e.g., log, Yeo-Johnson, etc.).
- Other features that weren't transformed.
Normalization harmonizes their scales.

---

### **5. For Improved Numerical Stability**
Normalization can help mitigate numerical instability in calculations:
- Avoids issues with very large or very small values when computing distances or gradients.
- Improves model convergence during optimization.

---

### **6. When Aggregating Data from Different Sources**
If features originate from different sources with varying units or ranges (e.g., sensor data, financial metrics), normalization ensures:
- Consistency across features.
- Fair contribution of each feature.

---

### **7. For Visualization Purposes**
If you plan to visualize transformed data (e.g., in t-SNE, UMAP, or other dimensionality reduction techniques), normalization:
- Ensures features contribute equally to the visualization.
- Prevents misleading patterns due to magnitude differences.

---

### **8. In Deep Learning Pipelines**
- Deep learning models generally perform better when input features are normalized (e.g., pixel intensity values for images or input features for fully connected layers).
- L2 normalization is often used for embedding vectors (e.g., word embeddings) to ensure consistent length.

---

### **9. When Handling Outliers Post-Transformation**
- Sometimes, transformations like log or Yeo-Johnson reduce skew but don’t completely address outliers. Normalization (e.g., max normalization) can limit the influence of extreme values.

---

### **10. In Feature Engineering for Ensemble Models**
- When combining features for ensemble models (e.g., stacking, blending), normalization ensures all features contribute proportionately, regardless of their original scale.

---

### **11. In Scientific Applications Requiring Unit Norm**
- Certain domains (e.g., physics, chemistry) require normalized data to enforce unit vectors for computations like:
  - Force or energy calculations.
  - Directional or proportional analyses.

---

### **12. In Recommendation Systems**
- L2 normalization is often applied to user-item interaction matrices or embeddings for collaborative filtering to ensure similarity computations (e.g., cosine similarity) are meaningful.

---

### **13. When Using Pre-Trained Models**
- Pre-trained models (e.g., word embeddings, pre-trained neural networks) often expect input data to be normalized (e.g., image pixel values rescaled to [0, 1] or [-1, 1]).

---

### **Summary**
Normalization (L1, L2, or max) is needed **after transformations** when:
1. The algorithm's performance is sensitive to feature scales (e.g., distance-based models, regularization, PCA).
2. Features retain different scales despite transformations.
3. Sparse data or vectors need consistent magnitudes.
4. Aggregated or heterogeneous features need harmonization.
5. Ensuring numerical stability or fair contribution across features is critical.

By understanding these scenarios, you can decide when normalization is appropriate based on your specific data and model requirements.
